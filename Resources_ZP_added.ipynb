```python
#Bringing in Sales Data for comparison

location = "EV_sales_2015-2018.csv"
```

```python
ev_sales = pd.read_csv(location)
ev_sales_df = pd.DataFrame(ev_sales)
ev_sales_df
```

```python
most_sales_df = ev_sales_df
most_sales_df = most_sales_df.rename(columns = {"States" : "State"})
most_sales_df = most_sales_df.sort_values(by = 'EVs sold from 2015-2018', ascending = False)
most_sales_df.head(5)
```

```python
# Creating Average Mediam Household Income for each state to use in a scatter plot vs EV cars sold

for state in df_final['State']:
    ave_med_inc = ((df_final['Household Income 2015'] + df_final['Household Income 2016'] + df_final['Household Income 2017'] + df_final['Household Income 2018'])/4)
ave_med_inc
```

```python
df_no_pr = df_final.drop(index = 27, axis = 1)
df_no_pr
```

```python
#Adding new column to Data Frame
df_no_pr['Average Median Household Income'] = ave_med_inc
df_no_pr = df_no_pr.sort_values(by = 'State', ascending = True)
df_no_pr
```

```python
#Linear Regression
slope, intercept, r_value, p_value, std_err = st.linregress(df_no_pr['Average Median Household Income'],
                                                            ev_sales_df['EVs sold from 2015-2018'])
regress_vals = df_no_pr['Average Median Household Income'] * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))

plt.scatter(df_no_pr['Average Median Household Income'], ev_sales_df['EVs sold from 2015-2018'], alpha = .6,
           edgecolors = 'green', color = 'green')
plt.xlabel('Average Median Household Income')
plt.ylabel('EVs sold from 2015-2018')
plt.title('EVs Sold v Average Median Household Income by state')

plt.plot(df_no_pr['Average Median Household Income'], regress_vals,"r-")
plt.annotate(line_eq,(20, 20), fontsize=15,color="red")
```

California has more that 600,000 more EVs sold over the 4 year period that the next highest state (Florida).
This helps account for the upward trend in the regression.  Though through reference the emissions data, the upward trend in
EV sales still has not had a significant impact on CO2 emissions (thanks Texas).
