```python
# Dependencies
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import requests
from census import Census
from config import api_key
from functools import reduce


api_key = "3a577fa8be5affaef548e14483f7bc3338a99541"
c = Census(api_key, year=2015)
```

```python
census_data = c.acs5.state(("NAME","B19013_001E"), Census.ALL)
```

```python
census_pd = pd.DataFrame(census_data)
census_pd = census_pd.rename(columns={"B19013_001E": "Household Income 2015",
                                     "NAME": "State","state": "State ID"})
census_pd
```

```python
c = Census(api_key, year=2016)
census_data1 = c.acs5.state(("NAME", "B19013_001E"), Census.ALL)

census_pd1 = pd.DataFrame(census_data1)
census_pd1 = census_pd1.rename(columns={"B19013_001E": "Household Income 2016",
                                     "NAME": "State","state": "State ID"})
census_pd1
```

```python
c = Census(api_key, year=2017)
census_data2 = c.acs5.state(("NAME", "B19013_001E"), Census.ALL)

census_pd2 = pd.DataFrame(census_data2)
census_pd2 = census_pd2.rename(columns={"B19013_001E": "Household Income 2017",
                                     "NAME": "State","state": "State ID"})
census_pd2
```
