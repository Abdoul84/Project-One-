```python

# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
```

```python
emissionsbystate = "../Project-One-/CO2_by_StateCSV.csv"
emissionsyearly = "../Project-One-/emissionsyearlyCSV.csv"
evsales = "../Project-One-/All_EV_Sales_2015-2018.csv"
```

```python
year = pd.read_csv(emissionsyearly, sep=",")
year
```

```python
co2bystate = pd.read_csv(emissionsbystate, sep=",")
co2bystate["Yearly CO2 2015"] = 5371.8
co2bystate["Yearly CO2 2016"] = 5248.0
co2bystate["Yearly CO2 2017"] = 5207.8
co2bystate["Yearly CO2 2018"] = 5255.5


narrow_state = co2bystate[["State", "2015", "2016", "2017", "2018", "Yearly CO2 2015","Yearly CO2 2016","Yearly CO2 2017","Yearly CO2 2018"]]
narrow_state = narrow_state.drop(labels=range(51,57), axis =0)
```

```python
statepercent2015 = []
statepercent2016 = []
statepercent2017 = []
statepercent2018 = []

statepercent2015 =((narrow_state["2015"].astype(float))/(narrow_state["Yearly CO2 2015"].astype(float)))*100
statepercent2016 =((narrow_state["2016"].astype(float))/(narrow_state["Yearly CO2 2016"].astype(float)))*100
statepercent2017 =((narrow_state["2017"].astype(float))/(narrow_state["Yearly CO2 2017"].astype(float)))*100
statepercent2018 =((narrow_state["2018"].astype(float))/(narrow_state["Yearly CO2 2018"].astype(float)))*100

dictionary = {"State": narrow_state["State"],
             "Percent 2015": statepercent2015,
             "Percent 2016":statepercent2016,
             "Percent 2017":statepercent2017,
             "Percent 2018":statepercent2018}
percent = pd.DataFrame(dictionary)
percent = percent.set_index("State")
percent
```

```python
percentplot =percent.T.plot(grid = True, title = "Percentage of Carbon Emissons per State for years 2015-2018", style = ".-")
percentplot.set_xlabel("years")
percentplot.set_ylabel("Percentage of Carbon Emission")
percentplot.figure.savefig("Percentage_CO2_perState.png")
```

```python
percentage_change = []
percentage_change = (percent["Percent 2018"]-percent["Percent 2015"])
percentage_change_df = pd.DataFrame(percentage_change, columns=['Percentage Change of Emission 2018-2015'])
percentage_change_df = percentage_change_df.reset_index()
percentage_change_df
```

```python
summary = percentage_change_df["Percentage Change of Emission 2018-2015"].describe()
summary
```

```python
percentage_change = percentage_change_df.sort_values(by = 'Percentage Change of Emission 2018-2015', ascending = True)
percentage_change.to_csv("Percentage Change of Emission 2018-2015.csv")
percentage_change
```

```python
fig1, ax1 = plt.subplots()
ax1.set_title('Percentage Change of Emission from 2015 to 2018')
ax1.set_ylabel('Percentage Change of Emissions')
ax1.boxplot(percentage_change_df["Percentage Change of Emission 2018-2015"])
plt.show()
plt.savefig("Percentage_Boxplot")
```

```python
evsalesstate = pd.read_csv(evsales, sep = ",")
```

```python
evsalesstate.sort_values(by='EVs sold from 2015-2018', ascending= False)
```

```python
evsalesstate["EVs sold from 2015-2018"].describe()
```

```python
x = percentage_change_df["Percentage Change of Emission 2018-2015"]
y = evsalesstate["EVs sold from 2015-2018"]
types = evsalesstate['States']

fig, ax = plt.subplots()
ax.scatter(x, y)

ax.set_xlabel('Percentage Change of Emissions 2015-2018', fontsize=14)
ax.set_ylabel('EVs sold from 2015-2018', fontsize=8)
ax.set_title('EVs sold vs Percentage Change in Emissions', fontsize=18)

for i, txt in enumerate(types):
    ax.annotate(txt, (x[i], y[i]), xytext=(10,10), textcoords='offset points')
    plt.scatter(x, y, marker='x', color='red')
    
fig.savefig('EVs_Sold_Percentage_CO2_perState.png')
```
